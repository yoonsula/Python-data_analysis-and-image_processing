```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3]
y = [1,2,3]
plt.plot(x,y)
plt.title("My Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.savefig("Picture.png")
```

```python
x = np.linspace(0, np.pi*10, 500) #PI*10 너비에, 500개의 점을 균일하게 찍기
fig, axes = plt.subplots(2,1) #2개의 그래프가 들어가는 Figure 생성
axes[0].plot(x, np.sin(x)) #첫 번째 그래프는 사인(sin) 그래프
axes[1].plot(x, np.cos(x)) #두 번째 그래프는 코사인(cos) 그래프
fig.savefig("sin&cos.png")
```

선 그래프 그리기

```python
x = np.arange(-9,10)
y = x**2
#라인 스타일로는 '-',':','-.','--' 등이 사용될 수 있습니다.
plt.plot(x,y,linestyle=":",marker="*")
plt.show()
```

```python
x = np.arange(-9,10)
y1 = x**2
y2 = -x
plt.plot(x,y1,linestyle='-.',marker='*',color="red",label="y=x*x")
plt.plot(x,y2,linestyle=':',marker='o',color="blue",label="y=-x")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(shadow=True, borderpad=1) #shadow:음영효과, borderpad: 박스형태
plt.show()
```

```python
x = np.arange(-9,10)
y1 = x**2
plt.plot(x,y1,linestyle=":",marker="o",markersize=8,
        markerfacecolor="blue",markeredgecolor="red")
plt.show()
```

막대그래프 그리기

```python
x = np.arange(-9,10)
plt.bar(x,x**2)
plt.show()
```

누적막대 그리프 그리기

```python
x = np.random.rand(10)
y = np.random.rand(10)
z = np.random.rand(10)
data = [x,y,z]
x_array = np.arange(10)
for i in range(0,3):
    plt.bar(x_array, data[i], bottom = np.sum(data[:i], axis=0))
plt.show()
```

스캐터 그래프 그리기

```python
x = np.random.rand(10)
y = np.random.rand(10)
colors = np.random.randint(0,100,10)
sizes = np.pi*1000*np.random.rand(10)
plt.scatter(x,y,c=colors,s=sizes,alpha=0.7) #alpha: 투명도
plt.show()
```
